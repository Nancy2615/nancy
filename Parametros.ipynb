```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
tips
```

```python
sns.histplot(data=tips, x="tip",bins=15,cumulative=False)
plt.show()
```

```python
sns.displot(data=tips, x="total_bill")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill",y="tip")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill",y="tip",hue="sex")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill",hue="sex",stat="count"
,multiple="dodge")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill",hue="sex",kind="kde"
,legend=True,palette="dark",alpha=.5)
plt.show()
```

```python
sns.kdeplot(data=tips,x="tip",hue="sex",cumulative=True,shade=True)
plt.show()
```

```python
sns.kdeplot(data=tips,x="tip",hue="sex",cumulative=True,shade=True,bw_adjust=1)
plt.show()
```

```python
sns.kdeplot(data=tips,x="tip",hue="sex",cumulative=False,shade=True,bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips,x="tip",hue="sex",stat="count")
plt.show()
```

```python
sns.displot(data=tips,x="tip",hue="sex",kind="hist",multiple="stack")
plt.show()
```
